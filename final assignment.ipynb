```python
# pull libraries for visual representation and data manipulation

import pandas as pd  # Load the Pandas libraries with alias 'pd'
import numpy as np  # Load the numpy libraries with alias 'np'
import matplotlib as mpl  # load the matplotlib as mpl
import matplotlib.pyplot as plt  # Load the matplotlib libraries with alias 'plt'

# read data from file name survey.csv
pd.set_option('display.max_columns', 5000)
data = pd.read_csv("C:/Users/shuha/Documents/DATA1202/survey.csv")
#print the read file just to double check I got the file I intended to find
print (data)
```

```python
numCountries = data.groupby(['Country'])['Country'].count().reset_index(name = "total number appeared in list")

numCountries.sort_values("total number appeared in list", ascending = False, inplace = True,
kind = 'quicksort', na_position ='last')

numCountries.head(47)
```

```python
# generally from the headers available, the most common links for mental health issues are assosiated
# with gender,family history,treatment, work interference, seek help, mental health consequences. 
# since I am more interested to know the results of male and female attributes when it comes to work interference,
# I am going to find the numbers assosiated with it using graphs in United States since we have the most available information
# of US
```

```python
data1 = data[data['Country']==('United States')]
data2 = data1[data1['Gender'] == ('Male')]

data2.head()
```

```python
maleWorkInterference = data2.groupby(['work_interfere'])['work_interfere'].count().reset_index(name = "work interference in males in US due to mental health issue")
maleWorkInterference.head()
```

```python
maleWorkInterference.plot(x = 'work_interfere',  kind ='barh', figsize=(15, 5), color='#86bf91', zorder=2, width=0.50)

plt.show()
```

```python
#for females with mental health issues and its effect on their work

data3 = data1[data1['Gender'] == ('Female')]

femaleWorkInterference = data3.groupby(['work_interfere'])['work_interfere'].count().reset_index(name = "work interference in females in US due to mental health issue")
femaleWorkInterference.head()
```

```python
data4 = data1[data1['Gender'] == ('M')]
capMWorkInterference = data4.groupby(['work_interfere'])['work_interfere'].count().reset_index(name = "work interference in 'M's in US due to mental health issue")
capMWorkInterference.head()
```

```python
data5 = data1[data1['Gender'] == ('m')]
smallMWorkInterference = data5.groupby(['work_interfere'])['work_interfere'].count().reset_index(name = "work interference in 'm's in US due to mental health issue")
smallMWorkInterference.head()
```

```python
# set width of bar
barWidth = 0.2


 
# set height of bar
bars1 = [64, 36, 47, 132] #male
bars2 = [8, 12, 14, 37] #female
bars3 = [19, 6, 12, 33] #M
bars4 = [3, 1, 2, 7] #m
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
 
# Make the plot
plt.bar(r1, bars1, color='#7f6d5f', width=barWidth, edgecolor='white', label='Male')
plt.bar(r2, bars2, color='BLUE', width=barWidth, edgecolor='white', label='Female')
plt.bar(r3, bars3, color='#2d7f5e', width=barWidth, edgecolor='white', label='M/Trans')
plt.bar(r4, bars4, color='RED', width=barWidth, edgecolor='white', label='m/ non binary')
 
# Add xticks on the middle of the group bars
plt.xlabel('group', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], ['Never', 'Often', 'Rarely', 'Sometimes'])
 
# Create legend & Show graphic

plt.legend()
plt.show()
```
